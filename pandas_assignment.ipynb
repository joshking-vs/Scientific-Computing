#outliers

```python
q1 = data['Weight'].quantile(0.25)
q3 = data['Weight'].quantile(0.75)
iqr = q3 - q1 
q1,q3,iqr
```

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
data_no_out = data[(data["Weight"] >= lower_bound) & (data["Weight"] <= upper_bound)]
data_no_out
```
